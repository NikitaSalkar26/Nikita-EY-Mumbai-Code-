```python
# Import necessary libraries
import pandas as pd
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Load the cleaned data
df_cleaned = pd.read_csv('cleaned_reviews.csv')

# Function to clean the text data
def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove special characters
    text = re.sub(r'\W', ' ', text)
    # Remove digits
    text = re.sub(r'\d', ' ', text)
    # Convert to lowercase
    text = text.lower()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return ' '.join(filtered_text)

# Function to tokenize the cleaned text
def tokenize_text(text):
    return word_tokenize(text)

# Apply the cleaning function to the review content
df_cleaned['cleaned_content'] = df_cleaned['content'].apply(clean_text)

# Apply the tokenizing function to the cleaned content
df_cleaned['tokenized_content'] = df_cleaned['cleaned_content'].apply(tokenize_text)

# Save the cleaned and tokenized text data to a new CSV file
df_cleaned.to_csv('cleaned_tokenized_reviews.csv', index=False)

# Print the cleaned and tokenized DataFrame
print(df_cleaned[['reviewId', 'userName', 'cleaned_content', 'tokenized_content', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at']])

# Display the cleaned and tokenized DataFrame
from IPython.display import display
display(df_cleaned[['reviewId', 'userName', 'cleaned_content', 'tokenized_content', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at']])
```

```python
import pandas as pd
from textblob import TextBlob

# cleaned and tokenized data from the CSV file
df_cleaned = pd.read_csv('cleaned_tokenized_reviews.csv')

#  TextBlob to determine the sentiment of a given piece of text
def analyze_sentiment(text):
    if isinstance(text, str):
        blob = TextBlob(text)
        sentiment_score = blob.sentiment.polarity
        if sentiment_score > 0:
            sentiment = 'POSITIVE'
        elif sentiment_score < 0:
            sentiment = 'NEGATIVE'
        else:
            sentiment = 'NEUTRAL'
        return sentiment, sentiment_score
    else:
        # Handle cases where text is not a string (e.g., NaN)
        return 'NEUTRAL', 0.0

# Applying sentiment analysis function to each review in DF
df_cleaned[['sentiment', 'sentiment_score']] = df_cleaned['cleaned_content'].apply(lambda x: pd.Series(analyze_sentiment(x)))

# Save DF
df_cleaned.to_csv('sentiment_analyzed_reviews.csv', index=False)

# Print out the DataFrame to see the sentiment analysis results
print(df_cleaned[['reviewId', 'userName', 'cleaned_content', 'sentiment', 'sentiment_score', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at']])

# Display 
from IPython.display import display
(df_cleaned[['reviewId', 'userName', 'cleaned_content', 'sentiment', 'sentiment_score', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at']])

```

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.decomposition import LatentDirichletAllocation
from IPython.display import display

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Load the sentiment analyzed reviews
df_sentiment = pd.read_csv('sentiment_analyzed_reviews.csv')

# Drop rows with NaN in 'cleaned_content'
df_sentiment = df_sentiment.dropna(subset=['cleaned_content'])

# Function to clean up the text
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)

# Clean up the reviews
df_sentiment['cleaned_content'] = df_sentiment['cleaned_content'].apply(preprocess_text)

# Just throwing in some example topics
topic_words = {
    0: ["game", "dice", "play", "money", "win"],
    1: ["fun", "great", "love", "enjoy", "good"],
    2: ["time", "play", "friends", "family", "enjoy"],
    3: ["negative", "bad", "issue", "problem", "hate"],
    4: ["app", "update", "feature", "bug", "fix"]
}

# Show the top words for each topic
def display_topics(topic_words):
    topics = []
    for topic_idx, words in topic_words.items():
        topics.append(words)
        print(f"Topic {topic_idx}: {' '.join(words)}")
    return topics

no_top_words = 15
topics = display_topics(topic_words)

# Give the topics some real names
topic_labels = {
    0: "Game Play and Mechanics",
    1: "Positive Experiences and Fun",
    2: "Time with Friends and Family",
    3: "Negative Experiences and Issues",
    4: "App Features and Updates"
}

# Assign topics to reviews (just random for now)
np.random.seed(42)  # keep it consistent
df_sentiment['topic'] = np.random.choice(list(topic_labels.keys()), len(df_sentiment))
df_sentiment['topic_name'] = df_sentiment['topic'].map(topic_labels)

# Show the DataFrame as a nice table
display(df_sentiment[['content', 'cleaned_content', 'sentiment', 'sentiment_score', 'topic_name']])

# Summarize the number of reviews in each topic and their overall sentiment
topic_summary = df_sentiment.groupby('topic_name').agg(
    num_reviews=('reviewId', 'count'),
    avg_sentiment_score=('sentiment_score', 'mean')
).reset_index()

# Show the summary as a nice table
display(topic_summary)

# Print detailed reviews in each topic
for topic_name in topic_labels.values():
    print(f"\nTopic: {topic_name}")
    display(df_sentiment[df_sentiment['topic_name'] == topic_name][['content', 'sentiment', 'sentiment_score']])

```
